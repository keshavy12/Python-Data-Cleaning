```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv("healthcare_data_cleaning_dataset.csv")

# Q1. Missing Data Identification
print("=== Q1 Missing Values ===")
missing_values = df.isnull().sum()
missing_percentage = (df.isnull().sum()/len(df))*100
print(missing_values)
print("\nMissing Percentage:")
print(missing_percentage)

# Q2. Handling Missing Age
median_age = df["Age"].median()
df["Age"] = df["Age"].fillna(median_age)
print("\nMedian Age Used:", median_age)

# Q3. Handling Missing Treatment Cost
median_cost = df["Treatment_Cost"].median()
df["Treatment_Cost"] = df["Treatment_Cost"].fillna(median_cost)
print("Median Treatment Cost Used:", median_cost)

# Q4. Duplicate Records
before = len(df)
duplicates = df.duplicated().sum()
df = df.drop_duplicates()
after = len(df)

print("\nDuplicate Rows:", duplicates)
print("Rows Before:", before)
print("Rows After:", after)

# Q5. Invalid Age Values
invalid_age = df[(df["Age"] < 0) | (df["Age"] > 100)]
print("\nInvalid Age Records:", len(invalid_age))

df = df[(df["Age"] >= 0) & (df["Age"] <= 100)]

# Q6. Outlier Detection (IQR)
Q1 = df["Treatment_Cost"].quantile(0.25)
Q3 = df["Treatment_Cost"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df["Treatment_Cost"] < lower) | (df["Treatment_Cost"] > upper)]
print("Number of Outliers:", len(outliers))

# Q7. Winsorization
p5 = df["Treatment_Cost"].quantile(0.05)
p95 = df["Treatment_Cost"].quantile(0.95)

df["Treatment_Cost_Capped"] = df["Treatment_Cost"].clip(lower=p5, upper=p95)

# Q8. Log Transformation
df["Log_Treatment_Cost"] = np.log1p(df["Treatment_Cost"])

plt.hist(df["Treatment_Cost"], bins=30)
plt.title("Before Log Transformation")
plt.show()

plt.hist(df["Log_Treatment_Cost"], bins=30)
plt.title("After Log Transformation")
plt.show()

# Q9. Time-Based Missing Handling
df["Admission_Date"] = pd.to_datetime(df["Admission_Date"])
df = df.sort_values("Admission_Date")
df = df.ffill()

print("\\nData Cleaning Completed Successfully")
```
